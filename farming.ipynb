```python
#importing of the libraries
import tensorflow as tf
import pandas as pd
import numpy as np
import math
```

```python
#preparing the data
#data preparation
Crop_recommendation= r'C:\Users\Mr.Green\Desktop\agri_Pro\Crop_recommendation.csv'
Crop_recommendation= pd.read_csv(Crop_recommendation)
print(Crop_recommendation.columns)
#print(Crop_recommendation.head())

```

```python
#creating independent and dependent vaeiables .....inputs and outputs
#dependent and independent data
independent = Crop_recommendation[['Nitrogen', 'Phosphorus', 'Potassium', 'temperature', 'humidity', 'ph', 'rainfall']]
dependent = pd.get_dummies(Crop_recommendation['crop'], prefix='')
print(independent.shape, dependent.shape)
```

```python
x = tf.keras.layers.Input(shape=[7])  # 7 input features
hidden = tf.keras.layers.Dense(8, activation='relu')(x)  # Hidden layer with 8 neurons
hidden = tf.keras.layers.Dense(12,activation='relu')(hidden)
hidden = tf.keras.layers.Dense(16,activation='relu')(hidden)
output = tf.keras.layers.Dense(22,activation='softmax')(hidden)
model = tf.keras.models.Model(x, output)
model.compile(loss='mse', optimizer='adam')

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Summary of the model
model.summary()
```

```python
#training the model we can use one of the two methods I chose the one which will show accuracy and loss
model.fit(independent,dependent,epochs=300,batch_size=32,verbose=1)
```

```python
test_loss, test_accuracy = model.evaluate(independent,dependent)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
model.save('mabala.keras')
```

```python
def predict_crops(model, Nitrogen,Phosphorus,Potassium,temperature,humidity,ph,rainfall):
    try:
        input_data = pd.DataFrame([[float(Nitrogen), float(Phosphorus), float(Potassium), float(temperature),float(humidity),float(ph),float(rainfall)]],
                                  columns=['Nitrogen','Phosphorus','Potassium','temperature','humidity','ph','rainfall'])

        # Scale input data
        input_data_scaled = scaler.transform(input_data)  # Scale the input data

        probabilities = model.predict(input_data_scaled)
        predicted_class = np.argmax(probabilities, axis=1)
        crop_map = {0: 'rice', 1: 'maize', 2: 'chickpea',3:'kidneybeans',4:'pigeonpeas',5:'mothbeans',6:'mungbean',7:'blackgram',8:'lentil',9:'pomegranate',10:'banana',11:'mango',12:'grapes',13:'watermelon',14:'muskmelon',15:'apple',16:'orange',17:'papaya',18:'coconut',19:'cotton',20:'jute',21:'coffee'}
        predicted_crop= crop_map[predicted_class[0]]

        return predicted_crop
    except Exception as e:
        print(f"Error in prediction: {e}")
        return "Error predicting crop"
    #input_data_scaled = scaler.transform(input_data)
```
